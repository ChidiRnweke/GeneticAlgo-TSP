```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
t500 = pd.read_csv("tour500results.csv")
t500ByTour = t500.groupby("Run").agg("mean")
t500ByMax = t500.groupby("Run").agg("max")
t500ByIeration = t500.groupby("# Iteration").agg("mean")

t500["Best value"].min()
```

```python
plt.figure(figsize=(12,6))
#plt.ylim(0,10000000)
plt.title("Mean of best values 4 runs of tour500.csv")
plt.plot( t500ByIeration["Best value"], label = "Best value")
#plt.plot(t500ByIeration["Mean value"], label = "Mean value")
plt.legend()
```

```python
plt.figure(figsize=(12,6))
#plt.ylim(0,10000000)
plt.title("Mean of mean values 4 runs of tour1000.csv")
#plt.plot( t500ByIeration["Best value"], label = "Best value")
plt.plot(t500ByIeration["Mean value"], label = "Mean value")
plt.legend()
```

```python
sns.displot(data= t500ByTour, x= "Best value")
plt.xticks( rotation='vertical');
```

```python
sns.displot(t500ByTour,  x = "Mean value")
plt.xticks( rotation='vertical');
```

```python
plt.figure(figsize=(12,6))
ax = sns.displot(t500ByMax,  x = "# Iteration")
plt.title("Amount of iterations until convergence")
plt.xticks( rotation='vertical');
```

```python
plt.figure(figsize=(12,6))

sns.displot(t500ByMax,  x = "Elapsed time")
plt.title("Amount of time until termination (seconds)")
plt.xticks( rotation='vertical');
```
